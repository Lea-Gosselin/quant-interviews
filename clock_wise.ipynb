# Clock pieces
$\sum_{n=1}^N=\frac{N(N+1)}{2}$

In this exercise, a clock (1-12) breaks into 3 pieces; the sum of the numbers on each pieces is equal.

```python
# Set variables
N=12 
p=3
```

```python
# The sum on each piece is...
s_um = ((N*(N+1))/2)/p
print("The sum on each piece is: ", s_um)
```

```python
# What are the number on all pieces?
def circular_range(start, length):
    return [(start + i - 1) % 12 + 1 for i in range(length)]

pieces = []
for i_start in range(1, 13):
    for piece_length in range(1, 13):
        piece = circular_range(i_start, piece_length)
        if sum(piece) == s_um:
            pieces.append(tuple(sorted(piece)))

# Delete duplicates
pieces = list(set(pieces))
used_numbers = {n for p in pieces for n in p}
pieces.append(tuple(sorted(set(range(1,N+1))-used_numbers)))

print("Possible pieces:")
for p in pieces:
    print(p)
```
