# Missing integers
We have $N$ integers within $[1,N]$. How can we easily find the $2$ missing integers?

```python
N=100

# We solve a system of 2 equations
def find_missing_numbers(nums, N):
    # Series summation
    S1 = N * (N + 1) // 2
    S2 = N * (N + 1) * (2 * N + 1) // 6

    # Missing numbers' summation
    m1 = sum(nums)
    m2 = sum(x * x for x in nums)

    # Summation and square summation
    xy_sum = S1 - m1
    xy_sq_sum = S2 - m2
    xy = (xy_sum * xy_sum - xy_sq_sum) // 2

    # Solve
    disc = xy_sum * xy_sum - 4 * xy
    x = (xy_sum + int(disc ** 0.5)) // 2
    y = (xy_sum - int(disc ** 0.5)) // 2

    return x, y

    
```

```python
m = [i for i in range(1,N+1) if i not in (37,87)]
find_missing_numbers(m, N)
```
